```python
import pandas as pd             
```

```python
df = pd.read_excel('Операторы.xlsx')
df
```

```python
df = df.groupby(['день', 'месяц']).agg(
    Count = pd.NamedAgg(column='минуты', aggfunc = 'sum')).reset_index()
df1 = df.copy()
df
```

```python
def func(Count):
    if Count <= 5:
        return Count*3.95
    else:
        return 5*3.95+(Count-5)*0.4
    
df['Стоимость'] = df.apply(lambda x: func(x.Count), axis = 1)
df
```

```python
df = df.groupby('месяц').agg(
    Paid = pd.NamedAgg(column='Стоимость', aggfunc = 'sum'))
df
```

```python
def func(Count):
    if Count <= 5:
        return Count*0.9
    elif Count <= 30:
        return 5*0.9+(Count - 5)*0.05
    else:
        return 5*0.9+25*0.05+(Count-30)*0.9
    
df1['Стоимость'] = df1.apply(lambda x: func(x.Count), axis = 1)
df1
```
