```python
import pandas as pd           
```

```python
df = pd.read_csv('Транзакции.csv', sep="\t", encoding = 'latin-1',decimal=",")
df
```

В таблице данные о 14635 транзакциях и 10 признаках. Переименуем столбцы с неправильной кодировкой. 

```python
df.rename(columns={ df.columns[4]: "Delivery", df.columns[5]: "Bad review" }, inplace = True)
df
```

Проверим данные на наличие пропущенных значений:

```python
df.isna().sum()
```

В таблице нет пропущенных значений. По условию задачи нас интересуют только те продажи, у которых не было доставки. Удалим строки, соответсвующие продажам с доставкой.

```python
df = df.drop(df[df.Delivery == 1].index)
df
```

Добавим дополнительный столбец, в котором будет отображаться только месяц и год.

```python
df['Date'] = pd.to_datetime(df['Date'])
df['year_month'] = df['Date'].dt.to_period('M')
df
```

Удалим ненужные столбцы.

```python
df = df.drop(['TransactionID', 'Date', 'Delivery', 'Weight', 'AutoIssue'], 1)
df
```

Сгруппируем данные по месяцу и промокоду, при этом данные из одной группы будут суммироваться, а для столбца плохих отзывов будет ещё считаться количество. 

```python
df = df.groupby(['year_month', 'Promo']).agg(
    BadReviewCount = pd.NamedAgg(column='Bad review', aggfunc = 'count'),
    BadReviewSum = pd.NamedAgg(column='Bad review', aggfunc = 'sum'),
    Count = pd.NamedAgg(column='Count', aggfunc = 'sum'),
    Cost = pd.NamedAgg(column='Cost', aggfunc = 'sum'),
    Paid = pd.NamedAgg(column='Paid', aggfunc = 'sum')).reset_index()
df
```

Таким образом, для каждого месяца есть две строки данных: с примененным промокодом и нет. Добавим ещё 2 столбца с вычисленными процентами плохих отзывов и доходности.

```python
def func1(BadReviewSum, BadReviewCount):
    return BadReviewSum/BadReviewCount*100

def func2(Cost, Paid):
    return (Paid - Cost)/Cost * 100

df['Bad review %'] = df.apply(lambda y: func1(y.BadReviewSum, y.BadReviewCount), axis=1)
df['Bad review %'].astype(int)
df['Profitability %'] = df.apply(lambda y: func2(y.Cost, y.Paid), axis=1)
df = df.round(decimals=2)
df    
```

Удалим ненужные для вывода столбцы. Переименуем нужные.

```python
df = df.drop(['BadReviewCount', 'BadReviewSum', 'Cost', 'Paid'], 1)
df.columns = ['Месяц','Введен промокод','Кол-во транзакций','Плохих отзывов %','Доходность, %']
df
```

В столбце промокода заменим 1 на "да", 0 на "нет".

```python
for i in range(0, df.shape[0]):
    if df.iloc[i, 1] == 1:
        df.iloc[i, 1] = 'да'
    else:
        df.iloc[i, 1] = 'нет'
df
```
